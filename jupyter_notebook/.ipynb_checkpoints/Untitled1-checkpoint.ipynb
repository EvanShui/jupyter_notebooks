```python
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool
import pandas_datareader as data
from bokeh.io import output_notebook
from datetime import date, timedelta
from dateutil.relativedelta import *
import numpy as np 
import pandas as pd
output_notebook()

delta_7_days = date.today() + relativedelta(days=-7)
delta_month = date.today() + relativedelta(months=-1)
delta_3_months = date.today() + relativedelta(months=-3)
delta_6_months = date.today() + relativedelta(months=-6)
delta_year = date.today() + relativedelta(years=-1)
delta_5_year = date.today() + relativedelta(years=-5)

start = datetime.datetime(2016,3,1)
end = datetime.datetime(2016,3,10) 

stock_ticker = "ATVI"
df = data.DataReader(name=stock_ticker,data_source="google",start=start,date.today())

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5, 6, 7, 8],
    price=np.array(df['Close'].values),
    desc=['A', 'b', 'C', 'd', 'E', 'F','G','H'],
))

hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("desc", "@desc"),
    ("price", "@price"),
    ], 
    mode='vline'
)

p = figure(x_axis_type='datetime', plot_width=400, plot_height=400, tools=[hover],
           title="Mouse over the dots")

p.line('x', 'price', size=20, source=source)

show(p)
```
