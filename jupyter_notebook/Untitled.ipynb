```python
from pandas_datareader import data, wb
import datetime 
from bokeh.plotting import figure,show,output_file, ColumnDataSource
from bokeh.io import output_notebook
from bokeh.models import HoverTool
from bokeh.layouts import widgetbox
from bokeh.models.widgets import RadioButtonGroup
from datetime import date, timedelta
from dateutil.relativedelta import *
import numpy as np
import pandas as pd

output_notebook()

delta_7_days = date.today() + relativedelta(days=-7)
delta_month = date.today() + relativedelta(months=-1)
delta_3_months = date.today() + relativedelta(months=-3)
delta_6_months = date.today() + relativedelta(months=-6)
delta_year = date.today() + relativedelta(years=-1)
delta_5_year = date.today() + relativedelta(years=-5)

start = datetime.datetime(2016,3,1)
source = None
stock_ticker = "ATVI"

hover = HoverTool(tooltips = [
        ("date", "@date{%F}"),
        ("Price", "$@price{0.2f}"),
        ("index", "$index")
    ],
    formatters = {
        "date":"datetime"
    },
    mode="vline"
)

p=figure(x_axis_type='datetime', width=500, height=200,
        title="Price vs. Day")
p.add_tools(hover)

def data_to_CDS(start_date):
    df = data.DataReader(name=stock_ticker,data_source="google",start=start_date,end=date.today())
    source = ColumnDataSource(data = dict(
        date=np.array(df['Close'].index, dtype=np.datetime64),
        price=np.array(df['Close'].values)
    ))
    return source

def update(new):
    print('Radio button option ' + str(new) + str(attr) + str(old) + ' selected.')
    if(new == "Option 1"):
        start = delta_7_days
        source = data_to_CDS(start)
        plot(source)
        show(widgetbox(radio_button_group))
        show(p)
    elif new=="Option 2":
        start = delta_month
        source = data_to_CDS(start)
        plot(source)
        
def plot(source):
    radio_button_group = RadioButtonGroup(labels=["Option 1", "Option 2", "Option 3"], active=0)
    radio_button_group.on_click(update)
    p.line('date', 'price', source=source)
```
